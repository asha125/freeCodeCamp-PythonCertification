```python
my_str = "Hello World"

uppercase_my_str = my_str.upper()
print(uppercase_my_str)

lowercase_my_str = my_str.lower()
print(lowercase_my_str)

my_str1 = "         hello world    " 
trimmed_my_str = my_str1.strip()
print(trimmed_my_str)


my_str2 = "hello world"
replaced_my_str = my_str2.replace("hello", "hi")
print(replaced_my_str)  

split_words = my_str.split()
print(split_words)


my_list = ["hello" , "world"]
joined_my_str = " ".join(my_list)
print(joined_my_str)

starts_with_Hello= my_str.startswith("Hello")
print(starts_with_Hello)

ends_with_world = my_str.endswith("world")
print(ends_with_world)

world_index = my_str.find("World")
print(world_index)

o_count = my_str.count("o")
print(o_count)

capitalized_my_str = my_str.capitalize()
print(capitalized_my_str)

is_all_upper = my_str.isupper()
print(is_all_upper)

is_all_lower = my_str.islower()
print(is_all_lower)

title_case_my_str = my_str.title()
print(title_case_my_str)
```
